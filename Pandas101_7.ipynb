## Plot - Bar - Histogram - Box 

```python
import pandas as pd; import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use("fivethirtyeight")
```

```python
data = pd.Series(np.random.randn(1000).cumsum())
data.plot()
```

```python
df = pd.DataFrame(np.random.randn(100,4),columns=list("ABCD"))
df = df.cumsum()
df.plot()
```

```python
df.iloc[10].plot(kind="bar")
```

```python
df.iloc[10].plot.bar()
```

```python
df1 = pd.DataFrame(np.random.rand(7,3),columns=list("ABC"))
df1.plot.bar()
```

```python
df1.plot.bar(stacked=True)
```

```python
df1.plot.barh(stacked=True)
```

```python
iris = pd.read_csv("C:\\Users\\erdem\\Desktop\\Data 101\\Pandas 101\\data\\iris.csv")
iris.head()
```

```python
iris.columns
```

```python
iris1 = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']]
iris1.plot.hist(alpha = 0.7)
```

```python
iris1.plot.hist(alpha = 0.7, stacked = True, bins = 25)
```

```python
iris1["SepalWidthCm"].plot.hist(orientation = "horizontal")
```

```python
iris1["SepalLengthCm"].diff().hist() # histogram of the differences of each value to the previous values
```

```python
plt.style.use("seaborn-whitegrid")
iris1.hist(color = "blue", alpha = 1, bins= 20)
```

```python
iris1.hist("PetalLengthCm", by = "Species")
```

```python
iris1.plot.box()
```

```python
color_ = {"boxes":"Red",
         "whiskers":"blue",
         "medians":"black",
         "caps":"green"}
iris1.plot.box(color = color_)
```

```python
iris1.plot.box(color = color_, vert = False)
```

```python
iris1.boxplot()
```

```python
plt.style.use("ggplot")
iris1.boxplot(by = "Species")
```

## Area - Scatter - Hexagonal - Pie - Density - Scatter Matrix

```python
df = pd.DataFrame(np.random.rand(10,4),columns=list("ABCD"))
df.head()
```

```python
df["A"].plot.area()
```

```python
df.plot.area(stacked= False)
```

```python
df.plot.scatter(x = "A", y = "B")
```

```python
iris1.head()
```

```python
iris1.plot.scatter(x = "SepalLengthCm", y = "SepalWidthCm")
```

```python
ax = iris1.plot.scatter(x = "SepalLengthCm", y = "SepalWidthCm", color = "Blue", label = "sepal")
iris1.plot.scatter(x = "PetalLengthCm", y = "PetalWidthCm", color = "Red", ax = ax)
```

```python
iris1.plot.scatter(x = "SepalLengthCm", y = "SepalWidthCm", c = "PetalLengthCm", s = 100)
```

```python
iris1.plot.scatter(x = "SepalLengthCm", y = "SepalWidthCm", s = iris1["PetalLengthCm"]*50)
```
